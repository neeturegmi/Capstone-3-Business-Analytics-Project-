# Capstone 2: Marketing Campaign

## Data Wrangling 

```python
# Import libraries 
import pandas as pd 
import numpy as np
```

```python
# load dataset 
df = pd.read_csv("/Users/rupashree/Downloads/marketing_campaign.csv", sep = '\t')
df.head()
```

```python
# Initial Inspection
print(df.shape)
print(df.info())
df.columns
```

```python
# Check missing values 
df.isnull().sum()
```

```python
# Fix income 
df['Income'].fillna(df['Income'].median(),inplace = True)
```

```python
df.info()
```

```python
# Feature Engineering 
df['Age'] = 2026 - df['Year_Birth']
```

```python
# Total Children 
df['Total_Children'] = df['Kidhome'] + df['Teenhome']
```

```python
# Total Spending 
df['Total_Spending'] = (
    df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] +
    df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
)
```

```python
# Customer tenure (days since enrollment)
# Convert Dt_Customer to datetime format first, then calculate the difference
df['Customer_Tenure'] = (pd.to_datetime('today') - pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')).dt.days
```

```python
#Check Income distribution 
df['Income'].describe()
```

```python
# Remove extreme outliers 
df = df[df['Income'] < 600000]
```

```python
# Check Unique values 
df['Marital_Status'].value_counts()
```

```python
# Simplify categories 
df['Marital_Status'] = df['Marital_Status'].replace({
    'Married': 'Couple',
    'Together': 'Couple',
    'Single': 'Single',
    'Divorced': 'Single',
    'Widow': 'Single',
    'YOLO': 'Other',
    'Alone': 'Single',
    'Absurd': 'Other'
})
```

```python
# Check duplicates 
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Final data check 
df.info()
df.describe()
```

```python
# Save clean dataset
df.to_csv("cleaned_marketing_data.csv", index=False)
```
